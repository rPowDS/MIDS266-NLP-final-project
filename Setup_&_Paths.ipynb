3.  **Project Directory Setup**: It sets up a base directory structure for the project at `/content/drive/MyDrive/w266_project_final` with subdirectories for data, models, outputs, configurations, tables, and figures. It also configures the Hugging Face cache to be stored within the project's data directory.
4. Due to the projects length and CPU demands this project was performed using my persoanl Google Drive to allow for CPU purchasing that was unaviable on the schoold Google Drive.

```python
# Mounting Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Base paths
import os
BASE_DIR = "/content/drive/MyDrive/w266_project_final"
DATA_DIR = os.path.join(BASE_DIR, "data")
MODELS_DIR = os.path.join(BASE_DIR, "models")
OUTPUTS_DIR = os.path.join(BASE_DIR, "outputs")
CONFIGS_DIR = os.path.join(BASE_DIR, "configs")
TABLES_DIR = os.path.join(BASE_DIR, "tables")
FIGS_DIR = os.path.join(BASE_DIR, "figs")

for d in [DATA_DIR, MODELS_DIR, OUTPUTS_DIR, CONFIGS_DIR, TABLES_DIR, FIGS_DIR]:
    os.makedirs(d, exist_ok=True)

# HF cache to project dir to speed up future runs
os.environ["HF_HOME"] = os.path.join(DATA_DIR, "hf_cache")

print("Base dir:", BASE_DIR)
```
